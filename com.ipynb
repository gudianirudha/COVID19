```python
corona_dataset_agreegated.loc["Spain"].diff().max()
```

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_agreegated.index)
max_infected_rate=[]
for c in countries:
    max_infected_rate.append(corona_dataset_agreegated.loc[c].diff().max())
```

```python
corona_dataset_agreegated["Max_Infected"] = max_infected_rate
```

### Task 3.4: create a new dataframe with only needed column 

```python
corona_dataset_agreegated.head()
```

```python
corona_data = pd.DataFrame(corona_dataset_agreegated["Max_Infected"])
corona_data
```
